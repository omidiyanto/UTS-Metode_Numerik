```python
#UTS Metode Numerik - O. Midiyanto (2001010021)
#soal No.7

#Persamaan f(x) No.7
def f(x):
    return -3*x**3 - 5*x + 9

#turunan persamaan f(x)
def g(x):
    return -9*x**2 - 5

x0=1.5  #nilai xn
x1=-1   #nilai xn+1
e=0.005 #nilai error


# program biseksi
def biseksi(x0, x1, e):
    print('\n\n*** METODE BISEKSI ***')
    print('NO       x0              x1           f(x0)         f(x1)           xmid        f(xmid)')
    iterasi = 1
    while True:
        x2 = (x0 + x1)/2
        print('%d       %0.6f      %0.6f      %0.6f     %0.6f        %0.6f     %0.6f' % (iterasi, x0, x1, f(x0), f(x1), x2, f(x2)))

        tanda_fx = ( f(x0) > 0 )
        tanda_fxmid = ( f(x2) > 0 )

        if (tanda_fx == tanda_fxmid):
            x0 = x2
        else:
            x1 = x2

        if( f(x2) > 0 ):
            mutlak_fxmid = f(x2)
        else:
            mutlak_fxmid = -1 * f(x2)

        if(mutlak_fxmid <= e):
            print('Akar Persamaannya ialah : %0.8f' % x2)
            break
        else:
            iterasi = iterasi + 1 


# program regula falsi
def regulaFalsi(x0, x1, e):
    print('\n\n*** METODE REGULA FALSI ***')
    print('NO       x0              x1           f(x0)         f(x1)           xmid        f(xmid)')
    iterasi = 1
    while True:
        x2 = x0 - (x1-x0) * f(x0)/(f(x1) - f(x0))
        print('%d       %0.6f      %0.6f      %0.6f     %0.6f        %0.6f     %0.6f' % (iterasi, x0, x1, f(x0), f(x1), x2, f(x2)))
       
        tanda_fx = ( f(x0) > 0 )
        tanda_fxmid = ( f(x2) > 0 )

        if (tanda_fx == tanda_fxmid):
            x0 = x2
        else:
            x1 = x2

        if( f(x2) > 0 ):
            mutlak_fxmid = f(x2)
        else:
            mutlak_fxmid = -1 * f(x2)

        if(mutlak_fxmid <= e):
            print('Akar Persamaannya ialah : %0.8f' % x2)
            break
        else:
            iterasi = iterasi + 1 


# program Newton Rhapson
def newtonRaphson(x0, e):
    print('\n\n*** METODE NEWTON RHAPSON ***')
    print('NO       x0              x1           f(x0)           f(x1)')

    iterasi = 1
    while True:
        x1 = x0 - f(x0)/g(x0)
        print('%d       %0.6f       %0.6f       %0.6f       %0.6f' %(iterasi, x0, x1, f(x0), f(x1)))
        
        if( f(x1) > 0 ):
            mutlak_fx1 = f(x1)
        else:
            mutlak_fx1 = -1 * f(x1)

        if(mutlak_fx1 <= e):
            print('Akar Persamaannya ialah : %0.8f' % x1)
            break
        else:
            iterasi = iterasi + 1 
            x0=x1

    
    
```
